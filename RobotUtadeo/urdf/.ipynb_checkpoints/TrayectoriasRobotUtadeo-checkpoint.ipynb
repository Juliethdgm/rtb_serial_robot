## Describir la trayectoria deseada del gripper para que el realice la letra M
### Letra M: Utilizamos SE3, e ingresamos las coordenas (x,y,z) en mm, asi con cada P# hasta realizar la letra M

### P es un punto en el plano cartesiano, es un punto donde quiero que mi efector final llegue 

### Nota: Esa posición debe estar dentro del espacio de trabajo 

```python
P1=SE3(x=100,y=300,z=100)
P2=SE3(x=100,y=300,z=520)
P3=SE3(x=240,y=300,z=260)
P4=SE3(x=380,y=300,z=520)
P5=SE3(x=380,y=300,z=100)
```

###  Se usa ctraj(T0, T1, n), ctraj es una trayectoria cartesiana desde la pose SE3 T0 hasta T1 con n puntos que siguen un perfil de velocidad trapezoidal a lo largo de la trayectoria. La trayectoria cartesiana es una instancia SE3 que contiene n valores.

### También usamos jtraj(q0, qf, N), jtraj es una trayectoria del espacio de la articulación donde las coordenadas de la articulación varían de q0(M) a qf(M), en N pasos.

```python
TrayCart1=rtb.ctraj(P1, P2, 100)
TrayCart2=rtb.ctraj(P2, P3, 100)
TrayCart3=rtb.ctraj(P3, P4, 100)
TrayCart4=rtb.ctraj(P4, P5, 100)
```

### Visualizar las trayectorias Cartesianas

```python
TrayCart1.plot()
TrayCart2.plot()
TrayCart3.plot()
TrayCart4.plot()
```

### Animación del robot haciendo las trayectorias
### Debemos pasar nuestras coordenadas de mm a cm, y usar nuevamente el método ctraj, para calcular las trayectorias cartesianas.

```python
P1=SE3(x=1.0,y=3.0,z=1.0)
P2=SE3(x=1.0,y=3.0,z=5.20)
P3=SE3(x=24.0,y=3.0,z=26.0)
P4=SE3(x=38.0,y=3.0,z=52.0)
P5=SE3(x=38.0,y=3.0,z=1.0)

TrayC1=rtb.ctraj(P1, P2, 50)
TrayC2=rtb.ctraj(P2, P3, 50)
TrayC3=rtb.ctraj(P3, P4, 50)
TrayC4=rtb.ctraj(P4, P5, 50)
```

### Ya teniendo las trayectorias cartesianas, calculamos la cinemática inversa, esto con el fin de obtener los valores que debe tomar cada articulación. 

```python
InvTrayC1=Robot.ikine_min(TrayC1)
InvTrayC2=Robot.ikine_min(TrayC2)
InvTrayC3=Robot.ikine_min(TrayC3)
InvTrayC4=Robot.ikine_min(TrayC4)
```

```python
print(InvTrayC2)
```

### Con los valores articulares obtenidos anteriormente, usamos el método jtraj, el cual calcula un polinomio de grado 5  para cada articulación que lleva de P_inicial a P_final, en un intervalo n

```python
TrayAr1 = rtb.jtraj(InvTrayC1[0].q, InvTrayC1[-1].q, 100)
TrayAr2 = rtb.jtraj(InvTrayC2[0].q, InvTrayC2[-1].q, 100)
TrayAr3 = rtb.jtraj(InvTrayC3[0].q, InvTrayC3[-1].q, 100)
TrayAr4 = rtb.jtraj(InvTrayC4[0].q, InvTrayC4[-1].q, 100)

#TrayAr1 = np.array([rtb.jtraj(InvTrayC1[0].q, InvTrayC1[-1].q, 100)])
#TrayAr2 = np.array([rtb.jtraj(InvTrayC2[0].q, InvTrayC2[-1].q, 100)])
#TrayAr3 = np.array([rtb.jtraj(InvTrayC3[0].q, InvTrayC3[-1].q, 100)])
#TrayAr4 = np.array([rtb.jtraj(InvTrayC4[0].q, InvTrayC4[-1].q, 100)])
#Array=np.vstack((TrayAr1,TrayAr2,TrayAr3,TrayAr4))
#print(Array)
```

### Animación del robot

```python
Robot.plot(TrayAr1.y,block=False)
time.sleep(1.9)

Robot.plot(TrayAr2.y,block=False)
time.sleep(1.9)

Robot.plot(TrayAr3.y,block=False)
time.sleep(1.9)

Robot.plot(TrayAr4.y,block=False)
```
