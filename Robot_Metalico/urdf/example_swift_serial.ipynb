```python
"""
This is the example to execute 
the Robot_Metalico
@author olmerg
"""
import sys  
sys.path.append( 'Robot_Metalico')
sys.path.append( '../../Swift_serial')
import numpy as np
import roboticstoolbox as rtb
from RobotMetalico import RobotMetalico
from Swift_serial import Swift_serial
from math import pi

if __name__ == '__main__':   # pragma nocover

    env = Swift_serial('COM9',115200)
    robot=RobotMetalico()
    print(robot)
    print(robot.to_dict)
    # the robot should start in home 
    env.launch()
    env.add(robot)    
#------------------------------------------------------------------------------------------------------------------------
    Tray1 = rtb.tools.trajectory.jtraj([0, 0, 0, 0, 0],[0, (2*pi/9), (2*pi/9), (-pi/3), 0],50)
    Tray2 = rtb.tools.trajectory.jtraj([0, (2*pi/9), (2*pi/9), (-pi/3), 0],[pi/2, (2*pi/9), (2*pi/9), (pi/3), 0],50)  
    Tray3 = rtb.tools.trajectory.jtraj([pi/2, (2*pi/9), (2*pi/9), (pi/3), 0],[0, 0, 0, 0, 0],50) 
    Tray4 = rtb.tools.trajectory.jtraj([0, 0, 0, 0, 0],[0, (2*pi/9), (2*pi/9), (-pi/3), 0],50)  
    Tray5 = rtb.tools.trajectory.jtraj([0, (2*pi/9), (2*pi/9), (-pi/3), 0],[-pi/2, (2*pi/9), (2*pi/9), (pi/3), 0],50) 
#------------------------------------------------------------------------------------------------------------------------    

#----------------------------
    for q in Tray1.y:
         #print(q)
         robot.q=q
         env.step(0.1)
#----------------------------
    for q in Tray2.y:
         #print(q)
         robot.q=q
         env.step(0.1)
#----------------------------
    for q in Tray3.y:
         #print(q)
         robot.q=q
         env.step(0.1)
#----------------------------
    for q in Tray4.y:
         #print(q)
         robot.q=q
         env.step(0.1)
#----------------------------
    for q in Tray5.y:
         #print(q)
         robot.q=q
         env.step(0.1)
#----------------------------

#----------------------------
    # return to home
    env.reset()
#----------------------------
```
